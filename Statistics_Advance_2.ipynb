# Statistics Advance 2 : Theorical Questions

1. What is hypothesis testing in statistics?
  - Hypothesis testing is a method used in statistics to make decisions or draw conclusions about a population based on a sample of data. It helps us determine whether there is enough evidence to support a certain belief or assumption (called a hypothesis) about a population parameter.
  - Hypothesis testing is a powerful tool for making data-driven decisions and testing ideas using evidence from samples.
  - There are two main types of hypotheses in hypothesis testing :
      1. Null Hypothesis (H₀)
      2. Alternative Hypothesis (Ha)

2. What is the null hypothesis and how does it differ from the alternative hypothesis?
  - In statistics, when we perform hypothesis testing, we start with two competing statements about a population : the null hypothesis and the alternative hypothesis.
  - Null Hypothesis (H₀) :
      - The null hypothesis is a statement that there is no effect, no difference or nothing new happening. It represents the status quo or a belief that any observed changes or results are due to random chance.
      - Represents the default or original claim.
      - It is the hypothesis that is tested directly.
      - Example : A medicine has no effect on patients.
      - Symbolically, it could be written as : H₀: μ = 100 (meaning the average is equal to 100).
  - Alternative Hypothesis (Ha) :
      - The alternative hypothesis is a statement that contradicts the null hypothesis. It suggests that there is a real effect, a difference or something new happening. This is the claim that researchers usually want to test or prove.
      - Represents a new claim or research hypothesis.
      - It is what we hope to support with evidence.
      - Example : The medicine does affect patients.
      - Symbolically : H₁: μ ≠ 100 (meaning the average is not equal to 100).


3. What is the significance level in hypothesis testing and why is it important?
  - The significance level, often denoted by the Greek letter alpha (α), is the threshold used in hypothesis testing to decide whether to reject the null hypothesis. It represents the probability of making a mistake by rejecting the null hypothesis when it is actually true - this mistake is called a Type I error.
  - Common Significance Levels :0.05 (5%) - most commonly used, 0.01 (1%), 0.10 (10%)
  - For example, if α = 0.05, it means we are willing to accept a 5% chance of wrongly rejecting the null hypothesis.
  - Why it is important :
      - It sets the standard for evidence needed to reject the null hypothesis.
      - Helps control the risk of making a Type I error (false positive).
      - A lower significance level (e.g., 0.01) means stricter criteria for claiming a result is statistically significant.
      - It helps researchers balance between being too cautious (missing real effects) and too risky (finding false positives).

4. What does a P-value represent in hypothesis testing?
  - The p-value is a number that helps us decide whether the results of our test are statistically significant.
  - The p-value tells us the probability of getting our observed results (or something more extreme) if the null hypothesis is actually true.
  - It shows how likely our data would occur just by random chance.
  - A small p-value means our results are unlikely to happen by chance, which suggests the null hypothesis may not be true.

5. How do you interpret the P-value in hypothesis testing?
  - The p-value helps us decide whether the results of our statistical test are significant or could have happened just by chance. Interpreting it correctly is key to understanding what our data is telling us.
  - Steps to Interpret the P-Value :
  1. Compare the p-value to the significance level (α) :
      - Common significance levels : 0.05, 0.01 or 0.10
  2. Make a decision :
      - If p <= α, The result is statistically significant. So, reject the null hypothesis.
      - If p > α, The result is not statistically significant. So, do not reject the null hypothesis (fail to reject the null hypothesis).
  - A small p-value (like 0.03) means our results are unlikely to happen by chance, so we likely have real evidence against the null hypothesis.
  - A large p-value (like 0.20) means our results could easily happen by chance, so there's not enough evidence to reject the null.

6. What are Type 1 and Type 2 errors in hypothesis testing?
  - In hypothesis testing, we make a decision about the null hypothesis (H₀) - we either reject it or do not reject it. But sometimes, we can make a mistake in our decision. That's where Type 1 and Type 2 errors come in.
  - Type 1 Error (False Positive) :
      - We reject the null hypothesis, but it was actually true.
      - We are saying there is an effect when there isn't one.
      - Example : We test a new medicine and say it works - but in reality, it doesn't. We made a Type 1 error. This is also called a "false alarm".
  - Type 2 Error (False Negative) :    
      - We do not reject the null hypothesis, but it was actually false.
      - We are saying there is no effect, but there actually is one.
      - Example : We test a new medicine and say it doesn't work - but in reality, it does. We made a Type 2 error. This is also called a "missed detection".

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
  - When we do hypothesis testing, we are trying to find out if there is a difference - but where we look for the difference depends on whether we are doing a one-tailed or a two-tailed test.
  - One-Tailed Test :
      - We check for a difference in only one direction.
      - We are asking : “Is it greater than or less than something?”
      - Example : We want to test if a new battery lasts longer than the old one.
      - H₀ (null) : New battery life = Old battery life
      - Ha (alternative) : New battery life > Old battery life
      - We only care if it's greater, not smaller.
      - Use this when : We are only interested in a result going in one specific direction.
  - Two-Tailed Test :
      - We check for a difference in both directions.
      - We are asking : “Is it different, whether it's higher or lower?”
      - Example : We want to test if a new teaching method gives different results (could be better or worse).
      - H₀ (null) : New method = Old method
      - Ha (alternative) : New method ≠ Old method
      - We care if it's better OR worse.
      - Use this when : We want to detect any change, in either direction.

8. What is the Z-test and when is it used in hypothesis testing?
  - A Z-test is a type of hypothesis test used to decide if there is a significant difference between sample data and a population (or between two samples), using the Z-statistic.
  - When we use a Z-Test - We use a Z-test when these conditions are true :
      1. The sample size is large (usually n ≥ 30).
      2. We know the population standard deviation (σ).
      3. The data is normally distributed or the sample is large enough for the Central Limit Theorem to apply.
  - Common Uses of Z-Test :
      - To compare a sample mean to a known population mean.
      - To compare two proportions from large samples.

9. How do you calculate the Z-score and what does it represent in hypothesis testing?
  -  The Z-score tells us how far our sample result is from the population mean, in terms of standard deviations.
  - Z-score = how "surprising" our sample result is compared to what we expected.
  - Calculate the Z-Score (Formula) :
      - For a sample mean, the formula is :
$$
Z = \frac{\bar{X} - \mu}{\frac{\sigma}{\sqrt{n}}}
$$
      - Where :
      - $\bar{X}$ = sample mean  
      - $\mu$ = population mean  
      - $\sigma$ = population standard deviation  
      - $n$ = sample size
  - The Z-score represents how many standard deviations our sample result is from the population mean, assuming the null hypothesis is true.
  - The Z-score tells us how far our result is from what we expected, if the null hypothesis is correct.
  - A high Z-score (positive or negative) means our result is very different from the null hypothesis - possibly significant.
  - A low Z-score means our result is close to expected, so likely not significant.



10. What is the T-distribution and when should it be used instead of the normal distribution?
  - The T-distribution is a type of probability distribution that is similar to the normal distribution (bell-shaped), but it has fatter tails.
  - These "fatter tails" mean there is more chance of values far from the mean, which helps when dealing with small sample sizes or unknown population standard deviation.
  - When we should use the T-distribution instead of the normal distribution :
      - Use the T-distribution when :  
      1. The sample size is small (usually n < 30).
      2. The population standard deviation is unknown.
      3. The data is approximately normally distributed.


11. What is the difference between a Z-test and a T-test?
  - Difference Between Z-Test and T-Test :
  - Sample Size :
      - Z-Test : Used when sample size is large (n ≥ 30).
      - T-Test : Used when sample size is small (n < 30).
  - Population Standard Deviation (σ) :
      - Z-Test : Known
      - T-Test : Unknown
  - Distribution Used :
      - Z-Test : Follows the normal distribution.
      - T-Test : Follows the t-distribution (has fatter tails).
  - Accuracy :
      - Z-Test : More accurate for large samples.
      - T-Test : More accurate for small samples.
  - Use Case :
      - Z-Test : When comparing sample mean to population mean with known σ.
      - T-Test : When comparing sample mean to population mean with unknown σ.

12. What is the T-test and how is it used in hypothesis testing?
  - A T-test is a statistical test used to find out if there is a significant difference between :
      - A sample mean and a population mean or
      - Two sample means.
  - It helps us decide whether the difference we see is real or just due to random chance.
  - How It's Used in Hypothesis Testing :
      1. State hypotheses :
      - Null hypothesis (H₀) : No difference
      - Alternative hypothesis (Ha) : There is a difference
      2. Choose significance level (α) :
      - Example : 0.05
      3. Calculate the T-statistic :
      - Use the sample data to compute the T-score.
      4. Find the critical value or p-value :
      - Based on degrees of freedom (df = n - 1)
      5. Make a decision :
      - If the T-score is extreme (or p-value < α), reject H₀.

13. What is the relationship between Z-test and T-test in hypothesis testing?
  - Relationship Between Z-test and T-test :
      - Both are hypothesis tests used to determine if there is a significant difference between means.
      - Both help us test assumptions about a population using sample data.
      - Main difference lies in sample size and known data :
          - Z-test is used when the sample size is large (n ≥ 30) and the population standard deviation is known.
          - T-test is used when the sample size is small (n < 30) and the population standard deviation is unknown.
      - Both calculate a test statistic (Z or T), which is compared to a critical value or used to find a p-value.
      - As the sample size increases, the T-distribution becomes closer to the normal (Z) distribution. This means : with large enough data, the T-test and Z-test give nearly the same result.
      - Both are used to determine whether to reject or fail to reject the null hypothesis.



14. What is a confidence interval and how is it used to interpret statistical results?
  - A confidence interval is a range of values that's likely to contain the true population value (like the mean), based on our sample data.
  - It gives us a range we can trust rather than a single guess.
  - For example : "We are 95% confident that the average score is between 68 and 72."
  - How to Use Confidence Intervals in Interpretation :    
      - A confidence interval gives us a range of likely values for the population parameter (like mean or proportion).
      - It is often written like this : Estimate ± Margin of Error
      - Example : 70 ± 2 → CI is [68, 72]
      - The confidence level (like 95%) tells how sure we are that the interval contains the true value.
          - 95% CI means: If we repeated the experiment many times, about 95 out of 100 intervals would include the true value.
      - Narrower CIs mean more precision; wider CIs mean more uncertainty.
      - In hypothesis testing, if the null hypothesis value is outside the confidence interval - we often reject H₀.

15. What is the margin of error and how does it affect the confidence interval?
  - The margin of error (MoE) is the maximum amount we expect our sample estimate to differ from the true population value.
  - It reflects the level of uncertainty in our estimate.
  - Formula (for mean, when population SD is known) :
$$
\text{Margin of Error} = Z \times \frac{\sigma}{\sqrt{n}}
$$

  (Or use the T-value when the sample size is small and σ is unknown.)
  - How It Affects the Confidence Interval (CI) :
      - The margin of error determines the width of the confidence interval :
          - Larger MoE - Wider confidence interval - more uncertainty.
          - Smaller MoE - Narrower confidence interval - more precision.
      - The confidence interval is calculated as : Sample Estimate ± Margin of Error
      - More confidence (e.g., 99% instead of 95%) - larger margin of error - wider interval.
      - Larger sample size - smaller margin of error - more accurate result.


16. How is Bayes' Theorem used in statistics and what is its significance?
  - Bayes' Theorem is a formula used to update probabilities based on new information or evidence.
  - It helps us revise what we believe (our prior probability) after seeing new data (called the likelihood).
  - Formula :    
$$
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
$$
  - What Each Term Means :
      - P(A|B) : Probability of A happening given B happened (posterior).
      - P(B|A) : Probability of B happening given A is true (likelihood).
      - P(A) : Initial belief or probability of A (prior).
      - P(B) : Total probability of B (evidence).
  - How it is used in statistics :
      - To update predictions or beliefs when new data comes in.
      - Common in :
          - Medical testing (e.g., updating disease probability after a test result).
          - Machine learning (like spam filters).
          - Risk analysis, decision-making under uncertainty.
  - Why it is important (Significance) :    
      - It shows how evidence changes our beliefs.
      - Unlike classical probability, which is fixed, Bayes' Theorem is dynamic - it evolves with data.
      - It forms the basis of Bayesian statistics, a powerful way of doing data analysis and prediction.

17. What is the Chi-square distribution and when is it used?
  - The Chi-square (χ²) distribution is a skewed, non-negative distribution (values are always ≥ 0).
  - The Chi-square distribution is used to test if observed data fits what we expect - especially useful for categorical data like surveys or experiments.
  - It is used to test how well observed data fits expected data.
  - Based on squared differences, so it's always positive.
  - When the Chi-square Distribution is used :
      - Chi-square Test for Independence :
          - Used to check if two categorical variables are related.
          - Example : Is gender related to preferred type of movie?
      - Chi-square Goodness-of-Fit Test :    
          - Used to check if observed values match expected values.
          - Example: Are dice rolls evenly distributed?
      - Used in variance analysis :
          - Also used in advanced statistics for testing population variance or in regression models.



18. What is the Chi-square goodness of fit test and how is it applied?
  - It is a statistical test used to check if observed data matches expected data for one categorical variable.
  - Use it when :
      - We have one categorical variable (e.g., color, brand, type).
      - We want to see if the distribution matches some expected pattern or proportion.
  - How it is applied :
  1. State hypotheses :
      - H₀ (null): The observed data matches the expected distribution.
      - H₁ (alternative): The observed data does not match the expected distribution.
  2. Collect observed data (O) :    
      - Actual counts we have recorded.
  3. Determine expected data (E) :
      - What we would expect if the null hypothesis were true.
  4. Use the Chi-square formula :
$$
\chi^2 = \sum \frac{(O - E)^2}{E}
$$

  5. Compare with the critical value (or calculate p-value) :
      - Use degrees of freedom = (number of categories - 1)
  6. Decision :
      - If χ² is large (or p-value < α), reject H₀
      - Otherwise, fail to reject H₀.

19. What is the F-distribution and when is it used in hypothesis testing?
  - The F-distribution is a probability distribution used mainly to compare variances between two groups.
  - It is positively skewed (values ≥ 0) and its shape depends on two parameters called degrees of freedom.
  - When the F-distribution is used in hypothesis testing - It's used in tests that involve comparing two or more variances. Most commonly in :    
  1. ANOVA (Analysis of Variance) :    
      - To test if three or more group means are significantly different.
      - Example : Comparing average scores of students from 3 different schools.
  2. F-test for comparing variances :    
      - To test if the variances of two populations are equal.
      - Example : Comparing the variability in heights of men vs women.


20. What is an ANOVA test and what are its assumptions?
  - ANOVA stands for Analysis of Variance.
  - It is a statistical test used to compare the means of three or more groups.
  - It tells us whether at least one group mean is significantly different from the others.
  - Example : Comparing the average test scores of students from 3 different schools.
  - Assumptions of ANOVA :
  1. Independence
      - The samples must be independent of each other.
  2. Normality
      - Data in each group should be roughly normally distributed.
  3. Homogeneity of variances
      - The groups should have equal variances (also called homoscedasticity).

21. What are the different types of ANOVA tests?
  - There are three main types of ANOVA tests, based on the number of factors and how they interact :    
  1. One-Way ANOVA :    
      - Used when comparing the means of 3 or more groups based on one factor.
      - Example : Comparing average test scores from 3 different schools (factor = school).
      - Checks only one independent variable.
  2.  Two-Way ANOVA :    
      - Used when comparing group means based on two independent variables (factors).
      - Example : Comparing test scores based on school and teaching method.
      - Can test for :
          - The individual effect of each factor (main effects).
          - The interaction effect between the two factors (do they affect each other?).
  3. Repeated Measures ANOVA :
      - Used when the same subjects are measured multiple times under different conditions.
      - Example : Measuring blood pressure before, during and after treatment for the same group of patients.
      - Accounts for within-subject variation.

22. What is the F-test and how does it relate to hypothesis testing?
  - The F-test is a statistical test used to compare two variances or to analyze multiple group means.
  - It is based on the F-distribution, which is a ratio of two variances.
  - How It Relates to Hypothesis Testing :
      - The F-test follows the standard hypothesis testing steps :
      1. Null hypothesis (H₀) : Variances (or means) are equal.
      2. Alternative hypothesis (H₁) : Variances (or at least one mean) is different.
      3. Calculate F-statistic.
      4. Compare with critical value from F-distribution table or use p-value.
      5. Decision :
          - If F is too large - Reject H₀
          - If F is small - Fail to reject H₀.

```python
#2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the range for the discrete uniform distribution
low = 1
high = 7  # upper bound is exclusive

# Create the distribution
dist = randint(low, high)

# Values and PMF
x = np.arange(low, high)
pmf = dist.pmf(x)

# Plot the PMF
plt.stem(x, pmf, basefmt=" ")
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('PMF of Discrete Uniform Distribution')
plt.grid(True)
plt.show()
```

```python
#3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.
def bernoulli_pmf(x, p):
    """
    Calculate the PMF of a Bernoulli distribution.

    Parameters:
    x (int): 0 or 1 (the outcome)
    p (float): probability of success (0 ≤ p ≤ 1)

    Returns:
    float: probability of the given outcome
    """
    if x not in [0, 1]:
        return 0  # x must be 0 or 1
    return p**x * (1 - p)**(1 - x)

print("P(X=0):", bernoulli_pmf(0, 0.3))
print("P(X=1):", bernoulli_pmf(1, 0.3))
```

```python
#5. Create a Poisson distribution and visualize it using Python.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Define the Poisson parameter (λ = average rate)
lam = 4

# Create the x values (possible event counts)
x = np.arange(0, 15)

# Get the PMF (Probability Mass Function)
pmf = poisson.pmf(x, mu=lam)

# Plot the distribution
plt.stem(x, pmf, basefmt=" ", linefmt='b-', markerfmt='bo')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.title('Poisson Distribution (λ = 4)')
plt.grid(True)
plt.show()
```

```python
#6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Range of the distribution (e.g., 1 to 6 like a die)
low = 1
high = 7  # upper bound is exclusive

# Create the discrete uniform distribution
dist = randint(low, high)

# Values to evaluate the CDF
x = np.arange(low, high)

# Calculate CDF values
cdf = dist.cdf(x)

# Plot the CDF
plt.step(x, cdf, where='post', color='purple', marker='o')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.title('CDF of Discrete Uniform Distribution (1 to 6)')
plt.grid(True)
plt.ylim(0, 1.1)
plt.show()
```

```python
#9. Write a Python function to calculate Z-scores from a dataset and plot them.
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std_dev

    # Plot Z-scores
    plt.plot(z_scores, marker='o', linestyle='-', color='purple')
    plt.axhline(0, color='black', linestyle='--', label='Mean (Z=0)')
    plt.title('Z-scores of Dataset')
    plt.xlabel('Data Index')
    plt.ylabel('Z-score')
    plt.grid(True)
    plt.legend()
    plt.show()

    return z_scores

data = np.array([10, 12, 23, 23, 16, 23, 21, 16])
z_scores = calculate_and_plot_z_scores(data)
print("Z-scores:", z_scores)
```

```python
#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def simple_error_viz(mu0, mu1, sigma, n, alpha=0.05):
    se = sigma / np.sqrt(n)
    z_crit = stats.norm.ppf(1 - alpha / 2)
    crit_low, crit_high = mu0 - z_crit * se, mu0 + z_crit * se

    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)
    y0 = stats.norm.pdf(x, mu0, se)  # under H0
    y1 = stats.norm.pdf(x, mu1, se)  # under H1

    plt.plot(x, y0, label='H0')
    plt.plot(x, y1, label='H1')

    plt.fill_between(x, y0, where=(x <= crit_low) | (x >= crit_high), color='red', alpha=0.3, label='Type I Error (α)')
    plt.fill_between(x, y1, where=(x > crit_low) & (x < crit_high), color='orange', alpha=0.3, label='Type II Error (β)')

    plt.axvline(crit_low, linestyle='--', color='black')
    plt.axvline(crit_high, linestyle='--', color='black')
    plt.legend()
    plt.title('Type I and Type II Errors')
    plt.show()

    beta = stats.norm.cdf((crit_high - mu1)/se) - stats.norm.cdf((crit_low - mu1)/se)
    print(f"Type I error (α): {alpha}")
    print(f"Type II error (β): {beta:.4f}")
    print(f"Power (1 - β): {1 - beta:.4f}")

simple_error_viz(mu0=100, mu1=105, sigma=15, n=50, alpha=0.05)
```

```python
#9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
    - data: list or array of sample values
    - confidence: confidence level (default = 0.95 for 95%)

    Returns:
    - (lower_bound, upper_bound)
    """
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # standard error of the mean
    margin = stats.t.ppf((1 + confidence) / 2, df=n-1) * std_err

    lower = mean - margin
    upper = mean + margin

    return (lower, upper)

sample = [88, 92, 85, 91, 87, 90, 89]
ci = confidence_interval(sample, confidence=0.95)

print(f"95% Confidence Interval for the mean: ({ci[0]:.2f}, {ci[1]:.2f})")
```

    - Narrower intervals indicate more precise estimates.
    - It's a key tool in reporting results of surveys, experiments and tests.

```python
#10. Write a Python program to calculate the margin of error for a given confidence level using sample data.
import numpy as np
import scipy.stats as stats

def margin_of_error(data, confidence=0.95):
    """
    Calculates the margin of error for a sample mean at a given confidence level.

    Parameters:
    - data: list or array of sample values
    - confidence: confidence level (default is 0.95 for 95%)

    Returns:
    - margin of error (float)
    """
    n = len(data)
    std_err = stats.sem(data)  # Standard Error of the Mean
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n - 1)  # Critical t-value

    margin = t_critical * std_err
    return margin

sample = [88, 92, 85, 91, 87, 90, 89]
moe = margin_of_error(sample, confidence=0.95)

print(f"Margin of Error (95% confidence): ±{moe:.2f}")
```

```python
#11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

# Bayesian Inference using Bayes' Theorem

# Suppose we want to determine the probability that a patient has a disease D given they tested positive.
# Given:
# P(D) = Prior probability of disease = 0.01 (1% of people have the disease)
# P(~D) = Probability of no disease = 1 - P(D)
# P(+|D) = Probability of testing positive given disease (true positive rate) = 0.99 (true positive rate)
# P(+|~D) = Probability of testing positive given no disease (false positive rate) = 0.05

P_D = 0.01            # Prior: 1% have the disease
P_not_D = 1 - P_D     # Complement: 99% do not have the disease

P_pos_given_D = 0.99      # True positive rate
P_pos_given_not_D = 0.05  # False positive rate

# Total probability of testing positive (evidence)
P_pos = (P_pos_given_D * P_D) + (P_pos_given_not_D * P_not_D)

# Apply Bayes' Theorem to compute posterior: P(D|+)
P_D_given_pos = (P_pos_given_D * P_D) / P_pos

# Bayesian Inference Result :
print(f"Probability of Disease given Positive Test (P(D|+)) = {P_D_given_pos:.4f}")
```

```python
#12. Perform a Chi-square test for independence between two categorical variables in Python.

# Example Scenario : We will use a fictional dataset showing the relationship between Gender and Preferred Beverage.
# We want to test if Gender and Beverage Preference are independent.

import pandas as pd
from scipy.stats import chi2_contingency

# Rows: Gender, Columns: Beverage Preferences
data = [
    [20, 15, 30],   # Male: Coffee, Tea, Water
    [25, 30, 15]    # Female: Coffee, Tea, Water
]

# Perform Chi-square test
chi2, p, dof, expected = chi2_contingency(data)

# Print the results
# Chi-Square Test Results
print(f"Chi-square Statistic = {chi2:.4f}")
print(f"Degrees of Freedom = {dof}")
print(f"P-value = {p:.4f}")
print("\nExpected Frequencies Table:")
print(pd.DataFrame(expected, columns=["Coffee", "Tea", "Water"], index=["Male", "Female"]))

# Interpret the result
alpha = 0.05  # significance level

if p < alpha:
    print("\nConclusion: Reject the null hypothesis.")
    print("-- There is a significant association between Gender and Beverage Preference.")
else:
    print("\nConclusion: Fail to reject the null hypothesis.")
    print("-- No significant association between Gender and Beverage Preference.")
```

```python
#13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.
import pandas as pd
from scipy.stats import chi2_contingency

# The observed data (contingency table)
data = [
    [30, 10],  # Self-study
    [20, 20]   # Group-study
]

observed_df = pd.DataFrame(data,
                           columns=["Pass", "Fail"],
                           index=["Self-study", "Group-study"])

print("Observed Frequencies :")
print(observed_df)

# chi2_contingency to get expected frequencies
chi2, p, dof, expected = chi2_contingency(observed_df)

# Convert expected output to a DataFrame for readability
expected_df = pd.DataFrame(expected,
                           columns=observed_df.columns,
                           index=observed_df.index)

print("\nExpected Frequencies :")
print(expected_df)
```

```python
#14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

# We want to test if the die is fair (i.e., each face has equal probability of 1/6).
from scipy.stats import chisquare

# Observed data from experiment
observed = [10, 9, 11, 10, 8, 12]  # Frequencies from 60 rolls of a die

# Define expected distribution (uniform if die is fair)
# Since 60 rolls and 6 sides => Expected = 60 / 6 = 10 per side
expected = [10, 10, 10, 10, 10, 10]

# Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Results
print("Chi-square Goodness-of-Fit Test")
print(f"Chi-square Statistic = {chi2_stat:.4f}")
print(f"P-value = {p_value:.4f}")

# Interpret the result
alpha = 0.05
if p_value < alpha:
    print("\nConclusion: Reject the null hypothesis.")
    print("-- The die is NOT fair (observed frequencies differ significantly).")
else:
    print("\nConclusion: Fail to reject the null hypothesis.")
    print("-- The die appears fair (no significant difference from expected).")
```
